# Data preprocessing

```python
import pandas as pd

# لازم تقرا الداتا الأول (دي كانت ناقصة عندك)
df = pd.read_csv("movies_dataset.csv")

# hn3ml elkalm elkol lowercase 3lshan nshel eldublicatat w noise w n3ml elmatching 
df["overview"] = df["overview"].astype(str).str.lower()
df["genres"] = df["genres"].astype(str).str.lower()
df["cast"] = df["cast"].astype(str).str.lower()
df["director"] = df["director"].astype(str).str.lower()

# elmodel bysht8l 3la elcombined column, fa hncombine elcolumns elly 3ndna fe column wa7da
df["combined"] = (
    df["overview"] + " " +
    df["genres"] + " " +
    df["cast"] + " " +
    df["director"]
)

# mafish 7aga mising bs lle7tyayat elmodel
df = df.dropna(subset=["combined"])

# mafish 7aga duplicate bs lle7tyayat elmodel
df = df.drop_duplicates(subset=["title"])

# hnreset elindex 3lshan elmodel yb2a 3ndo index mn 0 l7ad n-1 
df = df.reset_index(drop=True)
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words="english")

tfidf_matrix = tfidf.fit_transform(df["combined"])
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```


# Index

```python
indices = pd.Series(df.index, index=df["title"].str.lower()).drop_duplicates()
```

# Recommendation Function

```python
def recommend(title, top_n=5):
    title_lower = title.lower()

    if title_lower not in indices:
        return "Movie not found"

    idx = indices[title_lower]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]

    movie_indices = [i[0] for i in sim_scores]

    return df["title"].iloc[movie_indices]
```

# Ngrb elmodel
